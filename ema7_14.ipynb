```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt
```

```python
Stratege_ema = ta.Strategy(
    name = 'ema7_ema14',
    description = 'ema7_ema14',
    ta = [
        {'kind': 'ema', 'length': 7},
        {'kind': 'ema', 'length': 14}
])
```

```python
def func_EMA7_14(df):

    df.ta.strategy(Stratege_ema)

    def ema_strategy(df):
        if (df.EMA_7 > df.EMA_14):
            return True
        else:
            return False
            
    df['signal'] = df.apply(ema_strategy,axis=1)
    df_vbt = df.ta.tsignals(df.signal, asbool=True, append=True, trade_offset=1)
    df_port = vbt.Portfolio.from_signals(
        df.Open,
        entries = df_vbt.TS_Entries,
        exits = df_vbt.TS_Exits,
        freq = 'D',
        init_cash = 100,
        fees = 0.0025
    )

    return df_port
```

# BTC

```python
btc = pd.read_csv('datatwoyear/BTC-USD.csv', index_col='Date')
btc_port = func_EMA7_14(btc.copy())
```

```python
btc_port.plot().show()
```

```python
btc_port.stats()
```
