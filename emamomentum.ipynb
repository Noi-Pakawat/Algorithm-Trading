```python
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt
```

```python
Stratege_ema_rsi = ta.Strategy(
    name = 'ema + rsi',
    description = 'ema + rsi',
    ta = [
        {'kind': 'ema', 'length': 7},
        {'kind':'rsi', 'length': 14}
])
```

```python
def func_ema_rsi(df):

    df.ta.strategy(Stratege_ema_rsi)

    def ema_rsi_strategy(df):
        if (df.Close > df.EMA_7) and (df.RSI_14 > 50):
            return True
        else:
            return False
            
    df['signal'] = df.apply(ema_rsi_strategy,axis=1)
    df_vbt = df.ta.tsignals(df.signal, asbool=True, append=True, trade_offset=1)
    df_port = vbt.Portfolio.from_signals(
        df.Open,
        entries = df_vbt.TS_Entries,
        exits = df_vbt.TS_Exits,
        freq = 'D',
        init_cash = 100,
        fees = 0.0025
    )

    return df_port
```

# BTC

```python
btc = pd.read_csv('datatwoyear/BTC-USD.csv', index_col='Date')
btc_port = func_ema_rsi(btc.copy())
```

```python
btc_port.plot().show()
```

```python
btc_port.stats()
```
